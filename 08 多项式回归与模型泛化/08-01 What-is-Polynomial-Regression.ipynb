## 多项式回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(777)

x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
```

```python
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x, y)
plt.show()
```

### 先用线性回归试试

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, y_predict, color='r')
plt.show()
```

### 解决方案，添加一个特征

```python
# 把 X 的平方 作为一个新的特征
print(X.shape)
print((X**2).shape)
```

```python
X2 = np.hstack([X, X**2])
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
plt.plot(x, y_predict2, color='r')
plt.show()
```

```python
# 图是乱的，是因为我们的 x 不是从小到大排列的，排序一下再画就好了：
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r')
plt.show()
```
